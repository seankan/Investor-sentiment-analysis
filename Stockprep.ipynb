### Stock Data preparation

#### Trial

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("stock_price_original/000001.csv")
```

```python
data.columns
```

```python
data["close"][0:0+200]
```

```python
data["1_day_return"] = np.log(data['close'])-np.log(data['close'].shift(1))
data["1_day_return"] = data["1_day_return"].shift(-1)
data["1_week_return"] = np.log(data['close'])-np.log(data['close'].shift(5))
data["1_week_return"] = data["1_week_return"].shift(-5)
data["1_month_return"] = np.log(data['close'])-np.log(data['close'].shift(22))
data["1_month_return"] = data["1_month_return"].shift(-22)
data["1_day_sd"] = np.zeros
for row in range(len(data)):
    data["1_day_sd"][row] = np.std(a = data['close'][row:row+250], ddof=1)/((250)**(1/2))
    if data["1_day_sd"][row]==0:
        data["1_day_sd"][row]=data["1_day_sd"][row-1]
        
data["1_day_sharp"] = data["1_day_return"]/data["1_day_sd"]
data["1_week_sharp"] = data["1_week_return"]/(data["1_day_sd"]*(5)**(1/2))
data["1_month_sharp"] = data["1_month_return"]/(data["1_day_sd"]*(22)**(1/2))
```

```python
data.to_csv("stock_price_original/000001.csv")
```
